# Visual Stories

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("cleaned_loan_dataset.csv")
```

#### TOP 10 Highest Loan Amount

```python
top_loans = df.nlargest(10, 'loan_amount')

plt.figure(figsize=(10,5))
sns.barplot(x='loan_amount', y=top_loans.index, data=top_loans)

plt.title("Top 10 Highest Loan Amounts")
plt.show()
```

Risk Category Dstribution

```python
df['risk_category'] = pd.cut(
    df['credit_score'],
    bins=[0,650,750,850],
    labels=['High Risk','Medium Risk','Low Risk']
)

sns.countplot(x='risk_category', data=df)

plt.title("Customer Risk Categories")
plt.show()
```

Average Loan Amount By Purpose

```python
purpose_loan = df.groupby('loan_purpose')['loan_amount'].mean().sort_values()

purpose_loan.plot(kind='barh')

plt.title("Average Loan Amount by Purpose")
plt.show()
```

Repayament Rate by Employment status

```python
repayment = pd.crosstab(
    df['employment_status'],
    df['loan_paid_back']
)

repayment.plot(kind='bar')

plt.title("Repayment Rate by Employment Status")
plt.show()
```

Average credit score by education Level

```python
education_score = df.groupby(
    'education_level'
)['credit_score'].mean()

education_score.plot(kind='bar')

plt.title("Average Credit Score by Education Level")
plt.show()
```

High Risk Customer Analysis

```python
high_risk = df[
    (df['credit_score'] < 650) &
    (df['debt_to_income_ratio'] > 40)
]

print("High Risk Customers:", len(high_risk))
```

High-Risk vs Low-Risk Customers

```python
risk_counts = df['risk_category'].value_counts()

risk_counts.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Customer Risk Distribution")
plt.show()
```
