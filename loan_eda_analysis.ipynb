TARGET VARIABLE ANALYSIS

Loan Repayment Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='loan_paid_back', 
              hue='loan_paid_back',
              data=df,
              palette=['red','green'],
              legend=False
              )
plt.title("Loan Repayment Distribution")
plt.xlabel("Loan Paid Back")
plt.ylabel("Number of customers")
plt.show()
```

# explanation:
##### The target variable loan_paid_back indicates whether a customer repaid the loan or not.
##### 1 → Loan successfully repaid.
##### 0 → Loan not repaid (default).
##### The count plot visualizes the number of customers in each category and helps identify the balance between loan repayments and defaults.

AGE DISTRIBUTION

```python
sns.histplot(df['age'], bins=20, kde=True,color='royalblue',edgecolor='black',alpha=0.8)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number Of Borrowers")
plt.show()
```

ANNUAL INCOME DISTRIBUTION

```python
sns.histplot(df['annual_income'], bins=30, kde=True,color='green',edgecolor='black')
plt.title("Annual Income Distribution")
plt.xlabel("Annual income")
plt.ylabel("Number Of Customers")
plt.show()
```

CREDIT SCORE ANALYSIS

```python
sns.histplot(df['credit_score'], bins=30, kde=True)
plt.title("Credit Score Distribution")
plt.xlabel("Credit Score")
plt.ylabel("Number Of Customers")
plt.show()
```

LONE AMOUNT DISTRIBUTION

```python
sns.histplot(df['loan_amount'], bins=30)
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
plt.ylabel("Number of customers")
plt.show()
```

GENDER WISE LOAN REPAYMENT

```python
sns.countplot(
    x='gender',
    hue='loan_paid_back',
    data=df
)
plt.show()
```

EDUCATION LEVEL VS REPAYMENT

```python
sns.countplot(
    x='education_level',
    hue='loan_paid_back',
    data=df
)
plt.xticks(rotation=45)
plt.show()
```

EMPLOYEMENT STATUS VS REPAYMENT

```python
sns.countplot(
    x='employment_status',
    hue='loan_paid_back',
    data=df
)
plt.xticks(rotation=45)
plt.show()
```

LOAN PURPOSE ANALYSIS

```python
sns.countplot(y='loan_purpose',data=df)
plt.show()
```

CREDIT SCORE VS LOAN PAID BACK

```python
sns.boxplot(
    x='loan_paid_back',
    y='credit_score',
    data=df
)
plt.show()
```

DEBT-TO-INCOME RATIO ANALYSIS

```python
sns.boxplot(
    x='loan_paid_back',
    y='debt_to_income_ratio',
    data=df
)
plt.show()
```

LOAN AMOUNT VS REPAYMENT

```python
sns.boxplot(
    x='loan_paid_back',
    y='loan_amount',
    data=df
)
plt.show()
```

CORRELATION ANALYSIS

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(20,14))
sns.heatmap(corr, annot=True)
plt.show()
```

PAIR PLOT(IMPORTANT FEATURES)

```python
sns.pairplot(
    df[['annual_income',
        'credit_score',
        'loan_amount',
        'debt_to_income_ratio',
        'loan_paid_back']],
    hue='loan_paid_back'
)
plt.show()
```

## Expected EDA Findings:
#### Customers with higher credit scores are more likely to repay loans.
#### Lower debt-to-income ratios correspond to better repayment performance.
#### Stable employment status positively impacts repayment.
#### Higher annual income reduces default risk.
#### Certain loan purposes may have higher default rates.
#### Credit score is one of the strongest predictors of repayment.
